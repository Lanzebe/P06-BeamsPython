```python
BeamStructure = Beam(1)

Shigley = BeamLab(BeamStructure)
Shigley.AddLoadCase('Shigley_1')
Shigley.LoadCases['Shigley_1'].AddFixedSupport('a', Position = 0)
Shigley.LoadCases['Shigley_1'].AddForce('b', Position = 1,Magnitude = -1)

Shigley.SolveAll()
Shigley.PlotAll()
```
