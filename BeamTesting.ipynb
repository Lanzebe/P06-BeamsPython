```python
Beam1 = Beam(12)

Lab1 = BeamLab(Beam1)
Lab1.AddLoadCase('LC1')

Lab1.LoadCases[0].AddForce('F_a', 1000, 12)
Lab1.LoadCases[0].AddForce('F_b', -2000, 8)
Lab1.LoadCases[0].AddMoment('M_a', -500, 5)


Lab1.LoadCases[0].AddFixedSupport('Root', Position = 0)
Lab1.LoadCases[0].AddPinnedSupport('Tip', Position = 10)
#Lab1.LoadCases[0].AddVerticalSliderSupport('Center', Position = 6)
#Lab1.LoadCases[0].AddVerticalSliderSupport('Center', Position = 6.1)


#print(Lab1.LoadCases[0])
Lab1.LoadCases[0].SolveLoadCase()
```
